```python
# Extract number of items from the "File" column
df["Item Size"] = df["File"].str.extract(r'_(\d+)_items')[0].astype(int)

# Select only the relevant columns (Time columns for each algorithm and Capacity)
time_columns = [
    "Capacity",
    "DP-BottomUp Avg Time (ms)",
    "DP-TopDown Avg Time (ms)",
    "Greedy Avg Time (ms)",
    "GA Greedy Random Avg Time (ms)",
    "GA Greedy Population Avg Time (ms)"
]

df_filtered = df[time_columns + ["Item Size"]]

df_filtered.rename(columns={
    'DP-BottomUp Avg Time (ms)': 'DP-BottomUp',
    'DP-TopDown Avg Time (ms)': 'DP-TopDown',
    'Greedy Avg Time (ms)': 'Greedy',
    'GA Greedy Random Avg Time (ms)': 'GA Greedy Random',
    'GA Greedy Population Avg Time (ms)': 'GA Greedy Population'
}, inplace=True)
```

```python
import numpy as np
```

```python
def calculate_time_ratio(df, algorithm_column):
    time_ratios = []
    capacities = df['Capacity'].unique()

    # Loop through the different capacities and calculate the time ratio
    for i in range(1, len(capacities)):
        prev_capacity = capacities[i-1]
        curr_capacity = capacities[i]

        prev_time = df[df['Capacity'] == prev_capacity][algorithm_column].values[0]
        curr_time = df[df['Capacity'] == curr_capacity][algorithm_column].values[0]

        time_ratio = curr_time / prev_time
        time_ratios.append({'Algorithm': algorithm_column, 'From Capacity': prev_capacity,
                            'To Capacity': curr_capacity, 'Time Ratio': time_ratio})

    return pd.DataFrame(time_ratios)

def calculate_time_complexity(time_ratios_df):
    complexities = []

    for algorithm in time_ratios_df['Algorithm'].unique():
        alg_df = time_ratios_df[time_ratios_df['Algorithm'] == algorithm]
        exponents = []

        for _, row in alg_df.iterrows():
            scale_factor = row['To Capacity'] / row['From Capacity']
            time_ratio = row['Time Ratio']

            # Calculate the exponent k using the formula k = log(time_ratio) / log(scale_factor)
            if time_ratio > 0:  # Avoid division by zero or log of non-positive numbers
                k = np.log(time_ratio) / np.log(scale_factor)
                exponents.append(k)

        # Average the exponents to estimate overall time complexity
        avg_exponent = np.mean(exponents) if exponents else 0
        complexities.append({'Algorithm': algorithm, 'Estimated k': avg_exponent})

    return pd.DataFrame(complexities)
```

```python
algorithms = ['DP-BottomUp', 'DP-TopDown', 'Greedy', 'GA Greedy Random', 'GA Greedy Population']
time_ratios_df = pd.DataFrame()

# Calculate time ratios for each algorithm
for algorithm in algorithms:
    time_ratios_df = pd.concat([time_ratios_df, calculate_time_ratio(df_filtered, algorithm)], ignore_index=True)

# Calculate the estimated time complexity for each algorithm
complexity_df = calculate_time_complexity(time_ratios_df)

# Display the results
print("Time Ratios:")
print(time_ratios_df)
print("\nEstimated Time Complexities:")
print(complexity_df)
```

**Relationship:**

$$
\text{Time} \propto W^k
$$

**Formula for \(k\):**

$$
k = \frac{\ln(\text{Time Ratio})}{\ln(\text{Scale Factor})}
$$

```python
def plot_algorithm(df, algorithm_name):
    # Filter the DataFrame for the selected algorithm
    filtered_df = df[df["Algorithm"] == algorithm_name]

    sns.set_style("whitegrid")

    # Create the Catplot
    g = sns.catplot(x = "Item Size", y = "Avg Time (ms)", hue = "Capacity", data = filtered_df, kind = "bar",
                    height = 4, aspect = 1.2, palette = "Set1")

    # Customize the graph
    ax = g.ax
    ax.set_title(algorithm_name)
    ax.set_xlabel("Item Size")
    ax.set_ylabel("Avg Time (ms)")
    ax.margins(x=0.1)
    ax.set_yscale("log")  # Apply logarithmic scale to y-axis

    # Adjusting the y-axis limits based on the maximum value
    y_max = filtered_df["Avg Time (ms)"].max() * 1.1  # Add some margin to the max value
    y_min = filtered_df["Avg Time (ms)"].min()
    if y_min <= 0:
        y_min = 1e-1  # Avoid issues with log scale if minimum value is zero or negative

    ax.set_ylim(y_min, y_max)  # Set the y-axis limit

    # Adjust layout for better spacing
    g.fig.set_size_inches(10, 6)
    plt.subplots_adjust(bottom = 0.2, left = 0.1)
    plt.show()
```
